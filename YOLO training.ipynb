```python
import os
from ultralytics import YOLO

model_path = "D:\\new training\\yolov8n.pt" 
data_yaml = "D:\\new training\\dataset.yaml"  
epochs = 5  

save_path = r"D:\new training\runs\detect\train"
os.makedirs(save_path, exist_ok=True)

model = YOLO(model_path)
results = model.train(
    data=data_yaml,  
    epochs=epochs,   
    imgsz=640,       
    save_dir=save_path  
)
metrics = results.metrics
results.plot()  

print(f"Model trained successfully! The best model is saved at {save_path}\\weights\\best.pt")
```

```python
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the trained model
model = YOLO("D:\\new training\\runs\\detect\\train4\\weights\\best.pt")
test_images_path = "D:\\Frames\\vid5" 
results = model.predict(source=test_images_path, save=True, save_txt=False, save_conf=False)

# Visualize a few predictions
for idx, result in enumerate(results[:5]):  
    img_with_boxes = result.plot()  
    plt.figure(figsize=(10, 10))
    plt.imshow(img_with_boxes)
    plt.axis('off')
    plt.title(f"Test Image {idx + 1} - Predictions")
    plt.show()
```
